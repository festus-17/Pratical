```python
import requests as r
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv('Sports')
api_url = os.getenv('https://www.thesportsdb.com/api.php?ref=apilist.fun')

team_url = 'https://www.thesportsdb.com/api/v1/json/3/searchteams.php'

players_url = 'https://www.thesportsdb.com/api/v1/json/3/searchplayers.php'

# The parameters for the request (team name)
team_params = {'t': 'Arsenal'}
players_params = {'t': 'Arsenal'}  # You can change 'Arsenal' to any team you want to search

# Function to fetch team information
def get_team_info():
    # Make the GET request to the team search API
    response = r.get(team_url, params=team_params)

    if response.status_code == 200:
        data = response.json()

    # i didnt print the json file. its clumpsy please !

        # Check if 'teams' exists and is not empty
        if 'teams' in data and data['teams']:
            # Iterate through the teams list and print information
            for team in data['teams']:
                print(f"Team Name: {team.get('strTeam')}")
                print(f"Team ID: {team.get('idTeam')}")
                print(f"Country: {team.get('strCountry')}")
                print(f"Stadium: {team.get('strStadium')}")
                print(f"Description: {team.get('strDescriptionEN')}")
                print("="*40)
        else:
            print("No teams found.")
    else:
        print(f"Error fetching team data: {response.status_code}, {response.text}")

def get_team_players():
    # Make the GET request to the players search API
    response = r.get(players_url, params=players_params)

    # Check the status code and proceed if successful
    if response.status_code == 200:
        data = response.json()

        # i didnt print its clumpsy
        # Check if 'player' exists and is not empty
        if 'player' in data and data['player']:
            # Iterate through the players list and print player names
            for player in data['player']:
                print(f"Player Name: {player.get('strPlayer')}")
                print(f"Player Position: {player.get('strPosition')}")
                print(f"Player ID: {player.get('idPlayer')}")
                print("="*40)
        else:
            print("No players found for this team.")
    else:
        print(f"Error fetching players data: {response.status_code}, {response.text}")

def main():
    get_team_info()   # Fetch and display team information
    get_team_players() # Fetch and display players from the team

if __name__ == "__main__":
    main()
```

```python
from bs4 import BeautifulSoup as BS
import requests as r 

url = "https://www.billboard.com/pro/indie-labels-own-half-recorded-music-market-midia-report/"

# Send GET request to the webpage
response = r.get(url)
html_doc = response.text

# Parse the HTML content using BeautifulSoup
soup = BS(html_doc, 'html.parser')  # Specify parser for better consistency

# Find all <h1>, <h2>, <h3>... tags that contain article titles or section headers
# You can adjust this depending on which tags you want to extract
titles = soup.find_all(['h1', 'h2', 'h3', 'h4'])  # Example: Extract all headers

# Define filename for saving the titles
filename = 'blog.txt'

# Open the file in write mode ('w') to overwrite previous content
with open(filename, 'w') as file:
    for title in titles:
        # Write each title text to the file (if it exists)
        if title.text.strip():
            file.write(title.text.strip() + '\n')

print(f"Titles have been written to {filename}")
```

```python
import pandas as pd
import datetime as dt

data = pd.read_csv('titanic.csv')

average_age = data['Age'].mean()
max_age = data['Age'].max()
min_age = data['Age'].min()

average_fare = round(data['Fare'].mean(), 2)
max_fare = data['Fare'].max()
min_fare = data['Fare'].min()

current_date = dt.datetime.now().strftime('%Y-%m-%d')
current_time = dt.datetime.now().strftime('%I:%M:%S %p')

data['Average_Age'] = average_age
data['Max_Age'] = max_age
data['Min_Age'] = min_age

data['Average_Fare'] = average_fare
data['Max_Fare'] = max_fare
data['Min_Fare'] = min_fare

data['Date'] = current_date
data['Time'] = current_time

summary_columns = ['Average_Age', 'Max_Age', 'Min_Age', 'Average_Fare', 'Max_Fare', 'Min_Fare', 'Date', 'Time']
new_data = data[summary_columns]

new_data.to_csv('titanic_summary.csv', index=False)

read_data = pd.read_csv('titanic_summary.csv')
print(read_data.head())
```
